```python
# import symbolic capability to Python
from sympy import *
from sympy.abc import *
from sympy.plotting import plot
from IPython.display import display
# print things all pretty
init_printing()
```

# Size of the problem: n 

```python
n = symbols("n", positive=True, real=True)
```

# Different Functions f with different Time complexity

```python
f_constant = 20         #Here we define functions
f_linear = 4 * n + 10
f_linear2 = 10 * n +1
f_quadratic = 1/4 * n**2 + 1/2*n +1 
f_quadratic2 = 1/8 * n**2 
```

# Definition of Order of Growth (O) ("Big O")

Order of growth (O) is a set of functions whose asymptotic growth behavior is considered equivalent:

Two functions f(n) and g(n) have the equivalent order of growth if $\displaystyle \lim_{n \rightarrow \infty} \frac{f(n)}{g(n)} = c$ with $c > 0$ and $c < \infty $.


f(n) has a higher order of growth than g(n) if
$\displaystyle \lim_{n \rightarrow \infty} \frac{f(n)}{g(n)} = \infty $.

f(n) has a smaller order of growth than g(n) if $\displaystyle \lim_{n \rightarrow \infty} \frac{f(n)}{g(n)} = 0 $.

$f(n) \in O(g(n))$ if $\displaystyle \lim_{n \rightarrow \infty} \frac{f(n)}{g(n)} = c$ with $c < \infty $.

# Test Order of Growth with the Sympy limit function

```python
limit(f_quadratic / f_quadratic2, n, oo) #Its a constant therefore quadratic and quadric 2 have the same order of growth
```

```python
limit(f_quadratic / f_linear, n, oo) #Not belonging to same order of growth
```

```python
limit( f_linear2 / f_quadratic2, n, oo)
```

# Plotting Time Complexity

```python
expr_range = (n,0,30)
p = plot(
    f_constant,
    f_linear,
    f_quadratic,
    expr_range,
    show = False,
    legend = True
);

p[0].line_color = 'g'
p[1].line_color = 'b'
p[2].line_color = 'r'



p[0].label = 'constant'
p[1].label = 'linear'
p[2].label = 'quadratic'

p.show()
```

# Crossover point

```python
e1 = Eq(f_quadratic, f_linear) #Crossoverpoint where the quadratic function is higher than linear; Eq stands for equasion (Gleichung)
e1
```

```python
solve(e1, n)
```

```python
solve(e1, n)[0].evalf()  #evalf means we get it as float 
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
from sympy import symbols, Eq, log, factorial, solve
import sympy.plotting as syp
```

```python
# TODO: Define 
# - a logarithmic function
# - a log-linear (n log n) function
# - a cubic function
# - an exponential function
# - a factorial function

# Define the symbolic variable
n = symbols('n')

# Define the symbolic functions
logarithmic_func = log(n, 10)
log_linear_func = n * log(n)
cubic_func = n**3
exponential_func = 2**n  # Example with base 2
factorial_func = factorial(n)
```

```python
# TODO create multiple plots with different combinations of functions
# Define the range for plotting
expr_range = (n, 1, 10)  # Adjust range as needed

# Plot the functions
p = syp.plot(
    logarithmic_func,
    log_linear_func,
    cubic_func,
    exponential_func,
    factorial_func,
    expr_range,
    show=False,
    legend=True
)

# Customize the appearance of each function
p[0].line_color = 'g'
p[1].line_color = 'b'
p[2].line_color = 'r'
p[3].line_color = 'c'
p[4].line_color = 'm'

p[0].label = 'Logarithmic'
p[1].label = 'Log-linear (n log n)'
p[2].label = 'Cubic'
p[3].label = 'Exponential'
p[4].label = 'Factorial'

# Show the plot
p.show()
```

```python
# TODO create a log-log plot with your factorial, exponential, quadratic, log-linear, and linear function
# Define the range for plotting
x_vals = np.linspace(1, 10, 100)  # Adjust range as needed

# Calculate y values for each function
logarithmic_vals = [logarithmic_func.subs(n, x) for x in x_vals]
log_linear_vals = [log_linear_func.subs(n, x) for x in x_vals]
cubic_vals = [cubic_func.subs(n, x) for x in x_vals]
exponential_vals = [exponential_func.subs(n, x) for x in x_vals]
factorial_vals = [factorial_func.subs(n, x) for x in x_vals]

# Plotting on log-log scale
plt.figure(figsize=(10, 6))

plt.loglog(x_vals, logarithmic_vals, label='Logarithmic')
plt.loglog(x_vals, log_linear_vals, label='Log-linear (n log n)')
plt.loglog(x_vals, cubic_vals, label='Cubic')
plt.loglog(x_vals, exponential_vals, label='Exponential')
plt.loglog(x_vals, factorial_vals, label='Factorial')

plt.xlabel('log(n)')
plt.ylabel('log(y)')
plt.title('Log-Log Plot of Functions')
plt.legend()
plt.grid(True)
plt.show()
```
