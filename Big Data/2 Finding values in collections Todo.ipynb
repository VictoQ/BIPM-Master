# Find Value in Sorted list with Binary Search

Animation: https://yongdanielliang.github.io/animation/web/BinarySearchNew.html

```python
def find_sorted(items, value):
    """Binary search for a value in a list of sorted items
    Args:
        - items: sorted list of items
        - value: searched value
    Returns:
        - index of value or -1 if not in list
    """
    low = 0
    high = len(items) -1

    while low <= high:
        mid = low + (high - low) // 2

        if items[mid] == value:
            return mid
        elif items[mid] < value:
            low = mid + 1
        else:
            high = mid - 1

    return -1  # Value not found
```

```python
# Test the find_sorted function
sorted_items = sorted(items)  # Sorting the items list
index_found = find_sorted(sorted_items, value)

if index_found != -1:
    print(f"Value {value} found at index {index_found}")
else:
    print(f"Value {value} not found in the list")
```

### Question: 
* Time your function with differently sized inputs by
* searching for each input size for a non-existing element
* and plot the change of the time to the size of the input

(see 1 Timing Algorithms)

```python
def find_sorted(items, value):
    """Binary search for a value in a list of sorted items"""
    low = 0
    high = len(items) - 1 #Initializes variables 'low' and 'high' representung the lower and upper bounds of the search range in that list

    start_time = time.time()  # Start timing the execution of the binary search function

    while low <= high:      #Enters a loop to perform binary search until the search range is valid (i.e. 'low' is less than or equal to 'high')
        mid = low + (high - low) // 2 #Calculates the middle index within the current search range

        if items[mid] == value: #Check if the value at the middle index matches the target value.
            return mid

        elif items[mid] < value: #If the middle value is less than the target value, narrows the search range to the upper half by updating low to mid + 1
            low = mid + 1
        else:
            high = mid - 1 #If the middle value is greater than the target value, narrows the search range to the lower half by updating high to mid - 1

    return -1  # If the loop completed without finding the value, returns -1 to indicate that the value is not in the list

# Initialize lists to store input sizes and corresponding execution times
input_sizes = []
execution_times = []

# Vary the input size and time the function for each size
for n in range(100, 1001, 100):  # Input sizes from 100 to 1000 in steps of 100
    items = sorted([random.randint(0, n*10) for _ in range(n)])  # Sorted list of size n
    value = n + 1  # Value not in the list, worst-case scenario

    # Measure the execution time for the binary search function
    exec_times = []  # List to store execution times for multiple repetitions
    repetitions = 10000  # Number of repetitions for timing measurements

    for _ in range(repetitions):
        start_time = time.time()
        find_sorted(items, value)
        end_time = time.time()
        exec_time = end_time - start_time
        exec_times.append(exec_time)

    # Calculate the average execution time over all repetitions
    avg_exec_time = sum(exec_times) / repetitions

    input_sizes.append(n)
    execution_times.append(avg_exec_time)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(input_sizes, execution_times, marker='o')
plt.title('Time Complexity of Binary Search')
plt.xlabel('Input Size')
plt.ylabel('Execution Time (s)')
plt.grid(True)
plt.show()
```
