```python
# d) mean-function

# Get mean-value of domestic sales?
# movies_df['Domestic Sales (in $)'].mean()

# Task: show only movies that grossed more than the mean
mean_value = movies_df['Domestic Sales (in $)'].mean()
movies_df[movies_df['Domestic Sales (in $)'] > mean_value]
```

## 6. Dataframe Manipulation
* How to add new columns and augment/enhance available data for analysis?
* Create a new column: runtime in minutes
* Create a new column: sales per minute


```python
# task: we want to get the domestic sales per minute
movies_df.info()
```

```python
# problem: time is in a bad format
movies_df['Movie Runtime']
```

```python
# Convert time format
movies_df['Movie Runtime'].str.split(' ', expand = True)
# movies_df['Movie Runtime'].str.split(' ', expand = True)[2]
```

```python
# Add a new column to a dataframe - runtime in minutes
movies_df['Runtime in Minutes'] = (
                                    (movies_df['Movie Runtime']
                                         .str
                                         .split(' ', expand = True)[0]
                                         .astype(float) *60) 
                                        + 
                                    (movies_df['Movie Runtime']
                                         .str
                                         .split(' ', expand = True)[2]
                                         .astype(float))
                                    )
```

```python
# Add another column with sales per minute
movies_df['Sales per Minute'] = round(movies_df['Domestic Sales (in $)'] / movies_df['Runtime in Minutes'], 2)
```

```python
movies_df.sort_values(by=['Domestic Sales (in $)'], ascending = False)[:5]
```

```python
movies_df.sort_values(by=['Sales per Minute'], ascending=False)[:5]
```
