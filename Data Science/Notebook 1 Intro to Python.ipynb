## 4. Functions

* Functions are used to organize code in a more easy way (otherwise: one long document with hundreds of commands).
* Similar to a cookbook: in a pizza cookbook, you don't want to explain how to make the dough for each pizza-recipe. Instead, you would always refer to page x where readers can find the ultimate dough recipe.
* In code, it's the same: instead of writing down the same calculations hundreds of times, a function can be used to refer to the necessary calculation each time it needs to be used.
* Functions can take parameters as an input, do something with them and return an output.
* Code in Python needs to be organized with indentations to be executed correctly.

```python
def function_name(parameter1, parameter2="Default value"):
    ...
```

````python
# simple function for addition
def simple_addition(parameter1, parameter2):
    result = parameter1 + parameter2
    return result
````

## 6. Conditional Statements
- One of the core-functionalities in almost any programming language. 
- You can use conditional statements to formulate detailed logic that should follow a certain condition in your program.
- Conditions are similar to rules in a process, e.g., if a website visitor doesn't click on a advertisement 10 times, then we will show a different advertisement. 
- There are three main options for performing a comparison: greater than (>), smaller than (<) and equal to (==)

## 8. Loops

* To use the pizza-recipe analogy: let's assume you have an order for 10 pizzas in your restaurant: for each of these 10 pizzas you have to perform a set of operations:
  - Prepare the dough
  - Add tomato-sauce
  - Add cheese
  - and then (depending on the type of pizza) add more specific ingredients
* From a process-perspective, you would add 10 pizzas into a processing-pipeline where each pizza is prepared - one by one.
* This is similar in Python: We often want to process a set of many values, e.g., documents or numbers
* Loops can be used to iterate over a set of values, e.g. the contents of a list (see above)
* If you have a list of values, a loop helps to do something with each individual value in a list
* There are two general types: for-loops and while-loops
* As a rule of thumb: for-loops allow for more customization and while-loops simply loop until a certain condition is met

8.1 How can we use a funtion for every element in a list? Use the print-function to simply print every element in the following list:

````python
my_list = [1, 2, 3, 4, 5]

# Type your code here:
for x in my_list:
    print(x)
````

8.2 How can we create a list with a list-generator and print all values?

````python
my_list = range(0,4)

# Type your code here:
for value in my_list:
    print(value)
````

## 9. Looping over lists and working with items
How can we iterate over a list and sum up all the values?

````python
# your list
my_list = [1, 2, 3, 4]

# variable to store the result
summed_values = 0

# Task: Sum up all values in a list
for item in my_list:
    print(item)
    # add your code here:
    summed_values+=item

print(summed_values)
````
